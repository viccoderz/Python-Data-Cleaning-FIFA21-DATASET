so in the contract column
1. inconsistent data format
2. have to create 4 new columns
3.contract start

6.contract status

```python
#printing the values in contract cols which consist of on loan and free players
for index,row in fifa.iterrows():
  if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
    print(row['Contract'])
```

```python
def extract_contract_info(contract):
  if contract == 'Free' or 'On Loan' in contract:
    start_date = np.nan
    end_date = np.nan
    contract_length = 0
  else:
    start_date,end_date = contract.split(' ~ ')
    start_year = int(start_date[:4])
    end_year = int(end_date[:4])
    contract_length = end_year - start_year
  return start_date,end_date,contract_length
#assigning new cols
new_cols = ['Contract start','Contract end','Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
#applying func
for i in range(len(new_cols)):
  fifa.insert(loc = fifa.columns.get_loc('Contract')+1 +i,column = new_cols[i],value = new_data[i])
```

```python
fifa[['Contract','Contract start','Contract end','Contract Length(years)']].sample(5)
```

```python
#now adding column to check contract status
def contract_status(contract):
  if contract == 'On Loan':
    return 'On Loan'
  elif contract == 'Free':
    return 'Free'
  else:
    return 'Contract'
#making new col and applying the function
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1,'Contract Status',fifa['Contract'].apply(contract_status))
```

```python
fifa[['Contract','Contract start','Contract end','Contract Length(years)','Contract Status']].sample()
```

Here we can remove the cm string and assign the whole column as cm without mentioning,and there are feet and inches measurements mixed with the cm ,so we can convert it to **cm**

```python
def converting_cm(height):
  if 'cm' in height:
    return int(height.strip('cm'))
  else:
    feet,inches = height.split("'")
    total_inches = int(feet)*12 + int(inches.strip('"'))
    return round(total_inches*2.54)
#uppdating the height col
fifa['Height'] = fifa['Height'].apply(converting_cm)
```

```python
#renaming the height col to heigh in cm
fifa = fifa.rename(columns = {'Height':'Height(cm)'})
```

In this column there are kg and lbs weights mixed so we will rename the cols and store the values only in KG

```python
def converting_kg(weight):
  if 'kg' in weight:
    return int(weight.strip('kg'))
  else:
    pounds = int(weight.strip('lbs'))
    return round(pounds*0.453592)
#updating the col
fifa['Weight'] = fifa['Weight'].apply(converting_kg)
```

```python
#renaming to kg
fifa = fifa.rename(columns = {'Weight':'Weight(kg)'})
fifa['Weight(kg)'].unique()
```
